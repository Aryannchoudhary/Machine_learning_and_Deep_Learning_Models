```python
quotes=df['quote']
quotes.head()
```

```python
quotes=quotes.str.lower()
```

```python
import string
translator=str.maketrans('','',string.punctuation)
quotes=quotes.apply(lambda x:x.translate(translator))
```

```python
quotes.head()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
vocab_size=8978
tokenizer=Tokenizer(num_words=vocab_size)
tokenizer.fit_on_texts(quotes)
```

```python

word_index=tokenizer.word_index
print(len(word_index))
list(word_index.items())[:10]
```

```python
sequence=tokenizer.texts_to_sequences(quotes)
```

```python
for i in range(3):
  print(quotes[i])
```

```python
for i in range(3):
  print(sequence[i])
```

```python
X=[]
y=[]


for seq in sequence:
  for i in range(1,len(seq)):
    input_seq=seq[:i]
    output_seq=seq[i]
    X.append(input_seq)
    y.append(output_seq)
```

```python
len(X)
```

```python
len(y)
```

```python
max_len=max(len(x)for x in X)
print(max_len)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
X_padded=pad_sequences(X, maxlen=max_len, padding='pre')
```

```python
X_padded[0]
```

```python
y=np.array(y)
```

```python
y.shape
```

```python
X_padded.shape
```

```python
# one hot encoding on y
from tensorflow.keras.utils import to_categorical
y_one_hot=to_categorical(y, num_classes=vocab_size)
```

```python
y_one_hot.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN,LSTM, Dense
```

```python
embed_dim=50
rnn_units=128
```

RNN


```python
rnn_model=Sequential()
```

```python
rnn_model.add(
    Embedding(input_dim=vocab_size, output_dim=embed_dim, input_length=max_len)
)
rnn_model.add(
    SimpleRNN(rnn_units, return_sequences=True)
)
rnn_model.add(
    SimpleRNN(rnn_units)
)
rnn_model.add(
    Dense(vocab_size, activation='softmax')
)
```

```python
rnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
rnn_model.summary()
```

LSTM


```python
lstm_model=Sequential()
lstm_model.add(
    Embedding(input_dim=vocab_size, output_dim=embed_dim,
              input_length=max_len)
)
lstm_model.add(
    LSTM(rnn_units)
)

lstm_model.add(
    Dense(units=vocab_size, activation='softmax')
)
```

```python
lstm_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
lstm_model.summary()
```

```python
epochs=10
batch_size=128
```

```python
history_rnn=rnn_model.fit(
    X_padded, y_one_hot,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.2
)
```

```python
history_lstm=lstm_model.fit(
    X_padded, y_one_hot,
    epochs=100,
    batch_size=batch_size,
    validation_split=0.2
)
```

```python
from tensorflow.keras.models import load_model
lstm_model = load_model('lstm_model.h5')
```

```python
index_to_word = {}
for word, index in tokenizer.word_index.items():
    index_to_word[index] = word
```

```python
def predictor(model,tokenizer,text,max_len):
  text=text.lower()

  seq=tokenizer.texts_to_sequences([text])
  padded_seq=pad_sequences(seq,maxlen=max_len,padding='pre')
  pred=model.predict(padded_seq,verbose=0)
  pred_word_index=np.argmax(pred)
  return index_to_word[pred_word_index]
```

```python
seed_text="life is a"
next_word=predictor(lstm_model,tokenizer,seed_text,max_len)
print(seed_text,next_word)
```

```python
def generate_quotes(model,tokenizer,seed_text,max_len,num_words):
  for _ in range(num_words):
    next_word=predictor(model,tokenizer,seed_text,max_len)
    if next_word is None:
      break
    seed_text += " "+ next_word
  return seed_text
```

```python
seed="are you a"
generate_text = generate_quotes(lstm_model,tokenizer,seed,max_len,10)
print(generate_text)
```
